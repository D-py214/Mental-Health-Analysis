```python
mhd.drop('Patient Number', axis=1,inplace=True) 
# we are droping Patient Number columns as it will be in no use for us in making a predective model 
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder() 
# here we have imported LabelEncoder from sklearn library for converting categorical values to integer value
```

```python
def CCD(data,col): #CCD (Converting Categorical Data into numrical form so that our ML model can work efficiently )
    mhd[col]=LE.fit_transform(mhd[col]) #created a def function CCD for conversion 
    return mhd
```

```python
for i in ['Sadness', 'Euphoric', 'Exhausted', 'Sleep dissorder',
       'Mood Swing', 'Suicidal thoughts', 'Anorxia', 'Authority Respect',
       'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On',
       'Nervous Break-down', 'Admit Mistakes', 'Overthinking',
       'Sexual Activity', 'Concentration', 'Optimisim', 'Expert Diagnose']:
    CCD(mhd,i) # here we call out and run the above created def function 
```

```python
#Now we are goin to store our dataset's features in X and Y variable 
X=mhd.drop(columns=[ 'Expert Diagnose'])  # In X we store independant variable and we are going to drop target variable as it will be stored in Y variable 
y=mhd['Expert Diagnose']  # y variable has target variable 
```

```python
from sklearn.model_selection import train_test_split # again we are going to use sklearn library to import TrainTestSplit
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # here we have split the traing and testing data and given the size too 
```

```python
from sklearn.ensemble import RandomForestClassifier # we are going to use RandomForestClassifier to train our Model
```

```python
RFC= RandomForestClassifier(random_state=0) # obeject is created of RandomForestClassifier or easy use
RFC.fit(X_train, y_train) # here our model is training with help of training data
```

```python
# now we are going to check the score of our model so as know how efficient are our input variable to predict the outcomes
```

```python
from sklearn.metrics import accuracy_score # again we are importing from sklearn - accuracy score
```

```python
y_pred = RFC.predict(X_test) #here we got the predictions of model and we have stored in y_pred
accuracy_score(y_test, y_pred) # to check accuracy score we are giving actual and predicted vales 
accuracy_score(y_test, y_pred)*100 # and if we multiply it with 100 we get percentage of accuracy score 
```

```python
# so here we can our input variable are 87.5% efficient to predict the outcomes through Decision Tree model 
```

```python
from sklearn.metrics import classification_report # from sklearn we are importing classification_report to check the inall accuracy in different  metrics of classification
```

```python
print(classification_report(y_test, y_pred))
 #in print function we are using classification_report and giving actual and predicted values 
```
